```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import glob
```

```python
def get_Spectrum(folder, parameter):
    filepath = glob.glob(folder+"*.txt")
    file = [fp.split("/")[-1] for fp in filepath]
    
    Param = np.zeros(len(file))
    for (ii, f) in enumerate(file):
        fs = f[:-4].split("_")
        for x in fs:
            n = len(parameter)
            if x[0:n] == parameter:
                param = float(x[n:])
                break
        Param[ii] = param
    
    sort_order = np.argsort(Param)
    Param = np.array(Param)[sort_order]
    file = np.array(file)[sort_order]
    filepath = np.array(filepath)[sort_order]
    
    Energy = []
    
    for fp in filepath:
        Out = np.genfromtxt(fp)
        Energy.append(Out)

    N = max([len(E) for E in Energy])
    for n in range(len(Energy)):
        if len(Energy[n]) < N:
            Energy[n] = Energy[n]+[np.NaN]*(N-len(Energy[n]))
    Energy = np.array(Energy)

    return Param, Energy
```

```python
def get_Density(folder, parameter):
    filepath = glob.glob(folder+"*.out")
    file = [fp.split("/")[-1] for fp in filepath]
    
    Param = np.zeros(len(file))
    State = np.zeros(len(file))
    for (ii, f) in enumerate(file):
        fs = f[:-4].split("_")
        for x in fs:
            n = len(parameter)
            if x[0:n] == parameter:
                param = float(x[n:])
                break
        Param[ii] = param
        for x in fs[1:]:
            n = 1
            if x[0:n] == "E":
                state = int(x[n:])
                break
        State[ii] = state
    
    sort_order = np.lexsort((State, Param))
    Param = np.array(Param)[sort_order]
    State = np.array(State)[sort_order]
    file = np.array(file)[sort_order]
    filepath = np.array(filepath)[sort_order]
    
    Nb = []
    Nf = []
    
    for fp in filepath:
        Out = np.genfromtxt(fp)
        Nb.append(Out[:,0])
        Nf.append(Out[:,1])
    
    Nb = np.array(Nb)
    Nf = np.array(Nf)

    Density = {"Nb": Nb,
               "Nf": Nf}

    return Param, State, Density
```

# UBF infinite limit

```python
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (5,4)

for L in [4, 5, 6, 7, 8]:
    UBB, Energy = get_Spectrum("results/Data_L{}_NB{}_NF2_NBmax{}_t1.0_UBB_UBF0.0_UBFinftrue_pBtrue_pFfalse/".format(L,L,L), "UBB")

    fig, ax = plt.subplots()

    for ii in range(1, L):
        ax.plot(UBB, Energy[:,ii]-Energy[:,0], marker="o", color="black", markersize=3)

    ax.set_xlim(0, 10)
    ax.set_title("L={}".format(L))
    ax.set_yscale("log")
    ax.set_ylim(1e-5, 1e-0)
    plt.show()
```

```python
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (5,4)

ll = 1

for L in [4,5,6,7,8]:
    UBB, State, Density = get_Density("results/Data_L{}_NB{}_NF2_NBmax{}_t1.0_UBB_UBF0.0_UBFinftrue_pBtrue_pFfalse/".format(L,L,L), "UBB")

    UBB_ll = []
    Nb_ll = []
    Nf_ll = []
    for ii in range(len(UBB)):
        if State[ii] == ll:
            UBB_ll.append(UBB[ii])
            Nb_ll.append(Density["Nb"][ii])
            Nf_ll.append(Density["Nf"][ii])
    Nb_ll = np.array(Nb_ll)
    Nf_ll = np.array(Nf_ll)

    fig, axs = plt.subplots(2, layout="constrained", sharex=True)
    axs[0].pcolormesh(UBB_ll, range(1,L+1), Nb_ll.T, cmap="inferno")
    axs[1].pcolormesh(UBB_ll, range(1,L+1), Nf_ll.T, cmap="inferno")
    fig.suptitle("L={}".format(L))
    plt.show()
```

# Test polar coordinates

```python
plt.rcParams['figure.dpi'] = 150
plt.rcParams['figure.figsize'] = (5,4)

L = 5

Theta = np.linspace(0,45,10)
DeltaE = []
for theta in Theta:
    UBB = round(np.sin(np.pi*theta/180), 3)
    UBF = round(np.cos(np.pi*theta/180), 3)
    
    t, Energy = get_Spectrum("results/Data_L{}_NB{}_NF2_NBmax{}_t_UBB{}_UBF{}_UBFinffalse_pBtrue_pFfalse/".format(L,L,L,UBB,UBF), "t")
    DeltaE.append(Energy[:,1]-Energy[:,0])
    
    #fig, ax = plt.subplots()
    #for ii in range(L):
    #    ax.plot(t, Energy[:,ii]-Energy[:,0],"-", markersize=3)  
    #ax.set_xlim(107, 108)
    #ax.set_title("L={}".format(L))
    #ax.set_yscale("log")
    #ax.set_ylim(1e-5, 1e-0)
    #plt.show()

DeltaE = np.array(DeltaE)

fig, ax = plt.subplots()
im = ax.pcolormesh(t, np.tan(np.pi*Theta/180), DeltaE, cmap="inferno", norm=colors.LogNorm())
cbar = fig.colorbar(im, ax=ax)
ax.set_xlabel("$t$")
ax.set_ylabel(r"$U_{\rm BB}/U_{\rm BF}$")
plt.show()
```
